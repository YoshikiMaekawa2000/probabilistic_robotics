Ideal Robotクラスを追加（アニメーションの追加）

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.animation as anm  #アニメーション描画モジュール
matplotlib.use('nbagg')              #jupyter 上でアニメーションを再生する機能
import math
import matplotlib.patches as patches  #図形の描画モジュール
import numpy as np
```

```python
class World:
    def __init__(self, debug = False):   #デバッグ用のフラグを追加
        self.objects = []   #ここに、世界の中にあるrobotや障害物を登録
        self.debug = debug
        
    def append(self, obj):
        self.objects.append(obj)
        
    def draw(self):
        fig = plt.figure(figsize = (3,3)) #8*8 inchの図を用意
        ax = fig.add_subplot(111)         #サブプロットを準備
        ax.set_aspect('equal')            #縦横比を設定（ここでは１：１）
        ax.set_xlim(-5, 5)
        ax.set_ylim(-5, 5)
        ax.set_xlabel("X")
        ax.set_ylabel("Y")
        
        elems = []                #描画する図形のインスタンスのリスト
        
        if self.debug:               #デバッグ時はアニメーションを表示しない
            for i in range(1000):
                self.one_step(i, elems, ax)
        else:
            self.ani = anm.FuncAnimation(fig, self.one_step, fargs = (elems, ax), frames = 10, interval = 1000, repeat = False )
            #FuncAnimation:図、コールバック関数、コールバック関数の引数（２つ目以降の引数）、コールバック関数を呼び出す回数、アニメが切り替わる間隔（ミリ秒）
            plt.show()
            
    def one_step(self, i, elems, ax):
        while elems:                   #elemsを全削除
            elems.pop().remove()
            #pop:指定した位置の要素を削除し、その要素の値を取得（指定しない場合は末尾を削除）
            #remove:指定した値と同じ要素を検索し、その最初の要素を削除
            #ここremove()必要？？
            
        elems.append(ax.text(-4.4, 4.5, "t="+str(i), fontsize=10))
        for obj in self.objects:
            obj.draw(ax, elems)   
            if hasattr(obj, "one_step"):               #objが「one_step」属性を持つときにTrue
                obj.one_step(1.0)
        
            
        
        
```

```python
class IdealRobot:
    def __init__(self, pose, color = "black"):
        self.pose = pose  #姿勢の初期値（numpyの3次元配列）
        self.r = 0.2     #半径
        self.color = color
        self.poses = [pose]  #軌跡の描画用
        
    def draw(self, ax, elems):
        x, y, theta = self.pose
        x_n = x + self.r * math.cos(theta)
        y_n = y + self.r * math.sin(theta)  #それぞれの鼻先の座標
        
        elems += ax.plot([x, x_n], [y, y_n], color=self.color)     #線の描画
        c = patches.Circle(xy = (x, y), radius = self.r, fill=False, color = self.color)   #円の描画
        elems.append(ax.add_patch(c))    #patch（ここでは円）をsubplotに追加
        
        self.poses.append(self.pose)
        elems += ax.plot([e[0] for e in self.poses], [e[1]for e in self.poses], linewidth = 0.5, color="blue")
        
    @classmethod             #クラスをインスタンス化しなくても使用できる関数
    def state_transition(cls, nu, omega, time, pose):
        
```
